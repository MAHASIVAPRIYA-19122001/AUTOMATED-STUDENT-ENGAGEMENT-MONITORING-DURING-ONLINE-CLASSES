```python
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.models import load_model
import tensorflow as tf
from google.colab import drive
import os


model = load_model('/content/gdrive/MyDrive/model/planb.h5')


folder_path = '/content/gdrive/MyDrive/faces2/'

emotion_prob = []
emotion_pred = []

for i, filename in enumerate(os.listdir(folder_path)):
    if filename.endswith('.jpg'):
        # Load and preprocess the image
        img_path = os.path.join(folder_path, filename)
        img = image.load_img(img_path, target_size=(48, 48))
        img = np.expand_dims(img, axis=0)
        # Make a prediction using the loaded model
        result = model.predict(img)
        y_pred = np.argmax(result[0])
        print(f"{filename}: {result}, {filename}: {y_pred}")

        # Append emotion probabilities and predictions to corresponding lists
        emotion_prob.append(result[0])
        emotion_pred.append(y_pred)

        # If we have processed 60 frames, print the predicted class mode
        if (i + 1) % 60 == 0:
            set_num = (i // 60) + 1
            set_prob = np.array(emotion_prob[-60:])
            set_pred = np.array(emotion_pred[-60:])
            set_mode = np.argmax(np.sum(set_prob, axis=0))
            print(f"Set {set_num}: {set_pred}, Mode: {set_mode}")

# If there are any remaining frames, print the predicted class mode for those as well
if len(emotion_pred) % 60 != 0:
    set_num = (len(emotion_pred) // 60) + 1
    set_prob = np.array(emotion_prob[-(len(emotion_pred) % 60):])
    set_pred = np.array(emotion_pred[-(len(emotion_pred) % 60):])
    set_mode = np.argmax(np.sum(set_prob, axis=0))
    print(f"Set {set_num}: {set_pred}, Mode: {set_mode}")
```
